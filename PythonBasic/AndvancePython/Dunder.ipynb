```python
class Employee :
    def __init__(self,name,salary):
        self.name = name
        self.salary = salary
    
    def __str__(self):
        return f"The name is : {self.name} and my salary is.. {self.salary}"
    
    def __repr__(self):
        return f"Name : {self.name}\nSalary : {self.salary}"
    
    # def __len__(self):
    #     return  len(self.name)
    
e = Employee("Krish", 50000)
print(str(e))
print(repr(e))
print(len(e.name))
```
