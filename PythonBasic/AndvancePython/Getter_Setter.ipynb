```python
class Emp:
    def __init__(self,name,salary):
        self.name = name
        self._salary = salary
    @property # this is getter method
    def salary(self):
        return self._salary
    @salary.setter
    def salary(self,value):
        if value > 0 :
            self._salary = value
        else:
            print("Salary must be possitive ")
e = Emp("Krish Patel",50000)
print(e.salary)
e.salary = 70000
print(e.salary)

        
```
