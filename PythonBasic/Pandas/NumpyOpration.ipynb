```python
arr_int =np.array([10,20,30,40,50])
arr_float = np.array([1.5,2.3,3.8,1.2])
arr_mixed = np.array([10,2.5,"Krish",True])
print(arr_int)
print(arr_float)
print(arr_mixed)
```

```python
a= np.array([10,20,30,40,50])
b=np.array([1,2,3,4,6])
print("Add",a+b)
print("Sub",a-b)
print("Mul",a*b)
print("div",a/b)
print("Mean",np.mean[a])
print("Mean",np.mean[b])
```
