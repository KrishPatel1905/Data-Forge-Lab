```python
import ollama

# 1. Get AI answer
response = ollama.chat(
    model='tinyllama',
    messages=[
        {
            'role': 'user',
            'content': 'Why we need data science?'
        }
    ]
)

ai_answer = response['message']['content']
print("\nü§ñ AI Answer:\n")
print(ai_answer)

# 2. Ask user for rating
rating = int(input("\n‚≠ê Please give rating (1 to 5): "))

# 3. Generate review based on rating
def rating_to_review(rating):
    if rating == 5:
        return "Excellent explanation. Very clear and highly informative."
    elif rating == 4:
        return "Good explanation with useful information."
    elif rating == 3:
        return "Average answer, could be improved with more examples."
    elif rating == 2:
        return "Below average response, lacks clarity."
    elif rating == 1:
        return "Poor explanation and not very helpful."
    else:
        return "Invalid rating."

review = rating_to_review(rating)

# 4. Final Output
print("\nüìù Generated Review:")
print(review)
```
