```python
import json

def load_data(filename):
    file = open(filename, "r")
    data = json.load(file)
    file.close()
    return data


def find_people_you_may_know(user_id, data):

    # Store friends of each user..
    user_friends = {}

    for user in data["users"]:
        user_friends[user["id"]] = user["friends"]

    if user_id not in user_friends:
        return []

    direct_friends = user_friends[user_id]
    suggestions = {}

    # Check friends of friends
    for friend in direct_friends:
        for mutual in user_friends[friend]:

            if mutual != user_id and mutual not in direct_friends:

                if mutual in suggestions:
                    suggestions[mutual] = suggestions[mutual] + 1
                else:
                    suggestions[mutual] = 1

    # Convert suggestions dict to list
    result = []
    for person in suggestions:
        result.append(person)

    return result


data = load_data("data.json")
user_id = 1
recommendations = find_people_you_may_know(user_id, data)
print("People You May Know for User", user_id, ":", recommendations)
```
