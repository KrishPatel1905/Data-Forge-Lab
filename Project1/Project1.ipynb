```python
import json
```

```python
def Load(fname):
    with open(fname,"r") as file :
         data = json.load(file)
    return data
    
```

```python
data = Load("data.json")
display_Users(data)
```

```python
def display_Users(data):
    print("User Data and Their Connection")
    for user in data["users"] :
        print(f"{user['name']} (ID : {user['id']} - Friends {user['friends'] } and Like Pages{user['liked_pages']}")
    print("\nPages:\n")
    for page in data["pages"]:
        print(f"{page['id']}: {page['name']}")
        
       
```
