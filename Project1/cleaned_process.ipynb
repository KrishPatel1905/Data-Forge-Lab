```python
import json

def clean_data(data):

    #clean empty name .....
    new_users = []
    for user in data["users"]:
        if user['name'].strip()!="":
            new_users.append(user)
    data["users"] = new_users


    #clean duplicate  friends...
    for user in data["users"]:
        unique_friend = []
        for  friend in user["friends"]:
            if friend  not in unique_friend:
                unique_friend.append(friend)
        user["friends"] = unique_friend

    #clean Unactive user 
    active_user = []
    for user in data["users"]:
        if len(user['friends']) or len(user['liked_pages']) >0:
            active_user.append(user)
    data["users"] = active_user
    

    #remove duplicate id 
    unique_page = []
    page_id = []
    for page in data["pages"]:
        if page["id"] not in page_id:
            page_id.append(page["id"])
            unique_page.append(page)
    data["pages"] = unique_page
    return data 
    
            
        
        
data = json.load(open("data2.json"))
data = clean_data(data)
json.dump(data,open("clean_data.json","w"),indent = 4)
print("Data Cleaned Succesfully...")
```
